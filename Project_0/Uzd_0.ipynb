```python
def task_2():
   N = 5 # Change input here

   m1 = np.full((N, N), 3, dtype=int)

   m2 = np.arange(1, N+1)
   m2 = np.tile(m2, (N, 1))

   i, j = np.indices((N, N), dtype=int)
   m3 = (i + j) % 2

   m4 = np.array([[2, 0, 2], 
                  [0, 0, 0], 
                  [2, 0, 2]])
   k = int(np.ceil(N / 3))
   m4 = np.tile(m4, (k, k))[:N, :N]

   m12 = np.concat([m1, m2], 1)
   m34 = np.concat([m3, m4], 1)
   m1234 = np.concat([m12, m34], 0)

   m1234 = np.insert(m1234, N, 0, 1)
   m1234 = np.insert(m1234, N, 0, 0)

   print(m1234)

task_2()
```

```python
import time

def method_a(A, B, N, M):
   C = np.zeros((N, M), dtype=int)
   start_time = time.time()

   for i in range(N):
      for j in range(M):
         for k in range(N):
            C[i, j] += A[i, k] * B[k, j]

   end_time = time.time()
   return end_time - start_time


def method_b(A, B, N, M):
   C = np.zeros((N, M), dtype=int)
   start_time = time.time()

   for i in range(N):
      for j in range(M):
         C[i, j] = np.sum(np.multiply(A[i], (B)[:, j], dtype=int))

   end_time = time.time()
   return end_time - start_time


def method_c(A, B):
   start_time = time.time()

   C = np.dot(A, B)

   end_time = time.time()
   return end_time - start_time


def task_3(A, N):
   # N - Row size from 1st task
   M = 4
   # A - Matrix from 1st task

   B = np.full((N, M), 2, dtype=int)

   comp_time_a = method_a(A, B, N, M)
   comp_time_b = method_b(A, B, N, M)
   comp_time_c = method_c(A, B)

   print(f"Method a: {comp_time_a:.15f} seconds")
   print(f"Method b: {comp_time_b:.15f} seconds")
   print(f"Method c: {comp_time_c:.15f} seconds")


task_3(A, N)
```

```python
from matplotlib import pyplot as plt

def task_4():
   a = 0
   b = 5

   x = np.arange(a, b, 0.01)
   y = np.sin(5 * np.pi * x) + 10 * x

   c = 4
   y_c = np.sin(5 * np.pi * c) + 10 * c

   plt.plot(x, y, color='green', label='f(x)')
   plt.plot(c, y_c, 'o', color='red', label='c')

   plt.xlabel('x')
   plt.ylabel('y')
   plt.title('Func plot')
   plt.legend()

   plt.show()

task_4()
```

```python
def a(N):
   v = np.random.randint(-100, 100, N)
   return v[v & 1 == 0]


def b(N):
   v = np.random.randint(-100, 100, N)
   v[v < 0] = 0
   return v


def c(N):
   X = np.random.randint(-100, 100, size=(3, N))
   Y = np.random.randint(-100, 100, size=(3, N))

   return np.sqrt(np.pow(X[0] - Y[0], 2) + np.pow(X[1] - Y[1], 2) + np.pow(X[2] - Y[2], 2))


def d(N):
   v = np.random.randint(-100, 100, N)
   i = np.count_nonzero(v[:-1] > v[1:])
   ii = np.count_nonzero(np.sign(v[:-1]) != np.sign(v[1:]))

   return (i, ii)

def task_5():
   N = 10

   print(f"(a) {a(N)}")
   print(f"(b) {b(N)}")
   print(f"(c) {c(N)}")
   i, ii = d(N)
   print(f"(d) i = {i}\n    ii = {ii}")

task_5()
```
